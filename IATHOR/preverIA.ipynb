```python
import os.path

import tensorflow as tf
import numpy as np

RED   = '\033[91m'
GREEN = '\033[92m'
RESET = '\033[0m' 

IMG_HEIGHT = 224
IMG_WIDTH  = 224

MODEL      = tf.keras.models.load_model(r".\\model\\my_models_IA_THOR.keras")

CLASS_NAME = ['thor', 'randomphotos']

def predict_image(image_path, number):
    number +=1
   
    img = tf.keras.utils.load_img(
        image_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)

    predictions = MODEL.predict(img_array)
    score = tf.nn.sigmoid(predictions[0])

    if 100 * (1 - score[0]) > 100 * score[0]:
        print(
            f"\nthis image has {RED}{100 * (1 - score[0]):.2f}%{RESET} chance of being {CLASS_NAME[1].upper()}\nand a { 100* score[0]:.2f}% chance of being {CLASS_NAME[0].upper()} " )
    else:
        print(
            f"\nthis image has {GREEN}{100 * score[0]:.2f}%{RESET} chance of being {CLASS_NAME[0].upper()}\nand a {100 * (1 - score[0]):.2f}% chance of being {CLASS_NAME[1].upper()} ")

    if score > 0.5:
        print(f"{GREEN}Result {number}: It's your dog! (CONFIDENCE: {100* score.numpy()[0]:.2f}%{RESET}")
    else:
        print(f"{RED}Result {number}: It's not your dog, (CONFIDENCE: {100 * (1 - score.numpy()[0]):.2f}){RESET}")

path_the_test = r".\\imagem-teste"

list_photos = []
image_ext = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.webp')

if os.path.exists(path_the_test):
    for name in os.listdir(path_the_test):
        if name.lower().endswith(image_ext):
            list_photos.append(name)
    for number, photo in enumerate(list_photos):
        path = os.path.join(path_the_test, photo)
        print(photo.center(50,"-"))
        predict_image(path, number)
        print(f'End'.center(50, "-") + "\n")

else:
    print(f"ERROR: File {path_the_test} not found!")
```
