```python
import pandas as pd
import duckdb
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

def parquet_tratativa(caminho_arquivo: str, limite: int = 5) -> pd.DataFrame | None:
   
    caminho = Path(caminho_arquivo)
    if not caminho.exists():
        print(f"⚠️ Arquivo não encontrado: {caminho}")
        return None

    try:
        with duckdb.connect() as con:
            query = f"SELECT * FROM read_parquet('{caminho}') LIMIT {limite}"
            df = con.execute(query).df()
            
            # Configurações de exibição
            pd.set_option('display.max_rows', None)
            pd.set_option('display.max_columns', None)
            pd.set_option('display.width', None)

            return df

    except Exception as e:
        print(f"❌ Erro ao ler o arquivo: {e}")
        return None


# Uso
df = parquet_tratativa(r'.\dados_parquet\dados_202510.parquet')

if df is not None:
    print(df.info())
    display(df.head().T)
    display(df['PRODUTO_ATUAL'].value_counts())
    print("Valores únicos:", df['PRODUTO_ATUAL'].nunique())
```
