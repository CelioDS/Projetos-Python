```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

def parquetTratativa(arquivo):
    try:
        df = pd.read_parquet(arquivo)
        pd.set_option('display.max_rows', None)
        pd.set_option('display.max_columns', None)
        #df.info()
        #df.head().T
        return df
    except FileNotFoundError:
        print("Erro sem arquivo!")
        return None   
    except Exception as e:
        print(f"ERROR {e}")
        return None  

df = parquetTratativa('.\dados_parquet\dados_202510.parquet')

if df is not None:
    #display(df.info())
    #display(df.head().T)
    #display(df.value_counts('PRODUTO_ATUAL'))
    display(df.nunique())
```

```python

# Configura o estilo do grfico
sns.set_style("whitegrid")

# 1. Agrupa os dados por canal de venda e soma a quantidade 'QT'
vendas_por_canal = df.groupby('NM_CANAL_VENDA')['QT'].sum().sort_values(ascending=False).head(10)

# 2. Cria a figura e o grfico
plt.figure(figsize=(5, 5))
sns.barplot(x=vendas_por_canal.index, y=vendas_por_canal.values, palette='viridis')

# 3. Adiciona ttulos e rtulos
plt.title('Top 10 Canais de Venda por Quantidade', fontsize=16)
plt.xlabel('Canal de Venda', fontsize=12)
plt.ylabel('Quantidade Total de Contratos', fontsize=12)
plt.xticks(rotation=45, ha='right') # Gira os rtulos para no sobrepor

# 4. Exibe o grfico
plt.tight_layout()
plt.show()
```

```python
# 1. Converte a coluna 'DT_VENDA' para o formato de data (passo crucial!)


# 2. Agrupa os dados por data e soma a quantidade 'QT'
vendas_por_dia = df.groupby('DT')['QT'].sum()

# 3. Cria a figura e o grfico de linhas
plt.figure(figsize=(5, 5))
plt.plot(vendas_por_dia.index, vendas_por_dia.values, marker='o', linestyle='-')

# 4. Adiciona ttulos e rtulos
plt.title('Evoluo da Quantidade de Vendas no Tempo', fontsize=16)
plt.xlabel('Data da Venda', fontsize=12)
plt.ylabel('Quantidade Total de Contratos', fontsize=12)
plt.grid(True)

# 5. Exibe o grfico
plt.tight_layout()
plt.show()
```

```python

vendas_por_cidade = df.groupby('NM_CIDADE')['QT'].sum().sort_values(ascending=False).head(100)

# 2. Cria a figura e o grfico de barras horizontais
plt.figure(figsize=(5, 16))
sns.barplot(x=vendas_por_cidade.values, y=vendas_por_cidade.index, palette='plasma', orient='h')

# 3. Adiciona ttulos e rtulos
plt.title('Top 10 Cidades por Quantidade de Contratos', fontsize=16)
plt.xlabel('Quantidade Total de Contratos', fontsize=12)
plt.ylabel('Cidade', fontsize=12)

# 4. Exibe o grfico
plt.tight_layout()
plt.show()
```
